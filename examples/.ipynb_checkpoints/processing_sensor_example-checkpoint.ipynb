# Examining the Contents of the SensorArray

Consider the file processesing_sensors.yaml. For each of the sensors located in this file, only the neccessary items for the operations conducted in this example are included. This is the **name**, and **id** to match input signals to sensors. The **signal unit** of the sensor, which specifies what units in the input signal the sensor reads in. The **measure_units** field, which specifies what units the sensors signal may be transformed to, and finally the **equation** and **uncertainty** fields

In the **equation** field, for all of the thermocouples, the equations is specified as <typeK_thermocouple>. This is because the transformation from signal to measure units is a little more complicated than something that may be written out by hand in the file here. The function typeK_thermocouple is a python function stored in a conversion library. the brackets <function_name> let the sensor module know that it should look for the function in a specified library (specified as the path to a python module). Less complicated is the StaticPressureVenturi (a static pressure sensor) which has an arbitrarily picked linear form transforming the signal (x) to the measure units (y). The algabraic model of the sensor module attempts to determine if the equation may be solved explicitly in order to decrease compute time. Finally, the PressureDifferential sensor has a non-linear, implicit equation for the measure unit, which will be solved for by default using the secant method. 

In the **uncertainty** field, each of the thermocouples has a _list_, specified in .yaml format by the "-" dashes with the next line, indented after the field. The list contains an uncertainty (of 2.2) in the measure_units, and an uncertainty in %, and an operator to pick between the multiple supplied uncertainty calculation methods. The order is not restrictive in these elements, but units must be specified as they govern how the equations are applied. For the StaticPressureVenturi, and the PressureDifferential, the uncertainties are supplied as simply numbers with the associated units

The first step is to read in the sensor array

```python
import sys
sys.path.append('..')
from sensor.sensor import SensorArray
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

sensor_array = SensorArray.from_file('processing_sensors.yaml')
#for a more detailed example of how to report sensor data, look at the reporting_sensor
#_example notebook
table = sensor_array.to_table(headerfmt = 'id',columns = 'attribute',tablefmt = 'github',exclude = ['name'])
print(table)
```

Now, let's gain a sense of the data we will be working with. I've prepared some data artificially, and dumped it into a csv file that may easily be read into the pandas framework in a single command. This is typically not the case and a few more lines will usually be required in order to successfully read data into a dataframe from arbitarily stored data. We can see that all of the generated data has essentially the same form with some slight variations. 

The first plot on the left shows the temperature that has already been converted into [C], and the plot in the middle showns the unconverted temperature sensor data in [V]. Finally, the plot on the right shows the static pressure and differential pressure transducer data that has not yet been converted.

```python
df = pd.read_csv('artificial_data.csv',index_col = 0)
print(df)
fig,ax = plt.subplots(nrows = 1,ncols =3)
ax[0].plot(df[['T1 [C]','T2 [C]']])
ax[0].set_ylabel('Temperature [C]')
ax[1].plot(df[['T3 [V]','T4 [V]']])
ax[1].set_ylabel('Temperature Sensor Output [V]')
ax[2].plot(df[['DP1 [A]','P1 [A]']])
ax[2].set_ylabel('Temperature Sensor Output [A]')
```

This next step will demonsrate the first capability of the SensorArray for processing signal data: using the .convert() method

```python
converted_df = sensor_array.convert(df)

print(converted_df)
fig,ax = plt.subplots(nrows = 1,ncols =3)
ax[0].plot(converted_df[['T1 [C]','T2 [C]']])
ax[0].set_ylabel('Temperature [C]')
ax[1].plot(converted_df[['T3 [C]','T4 [C]']])
ax[1].set_ylabel('Temperature [C]')
ax[2].plot(converted_df[['DP1 [Pa]','P1 [Pa]']])
ax[2].set_ylabel('Pressure [Pa]')
```

We can see that the temperatures T1 and T2 have are untouched, as they have already been converted. However, T3 and T4 are converted according to the specified equation, likewise with the static pressure. It is important to note that the units are switched by the convert function in the headers of the data frame

Let's say we have incurred a warning regarding the DP1 conversion due to the nonlinear behaviour and the solve not being sufficient. How do we fix this issue with the pressure not being solve for correctly? We would specify a beefier root finder that is more computationally expensive but perhaps more robust _a la_ below where I have just specified the newton method with a greater number of iterations 

```python
from scipy.optimize import newton

x = np.linspace(-100,100,1000)
y = np.power((x+50.0),2)

fig,ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel('input signal [x]')
ax.set_ylabel('output of conversion [y]')

def root_finder(func,shape):
    
    return newton(func,np.zeros(shape),maxiter = 500) 


converted_df = sensor_array.convert(df,root_finder = root_finder)
print(converted_df)
```

The second capability we want to demonstrate involves computing the uncertainties associated with the sensors: using the .calculate_uncertainty() method of the sensor array. Similar to the .convert() method, the .calculate_uncertainty() method takes in a data frame $X$ with columns containing the signals of the sensors we wish to calculate the uncertainties for. 


```python
uncertainties = sensor_array.calculate_uncertainty(df)
print(uncertainties)
```
